Q1conduct a simulation study from N(ùúá,ùúé^2) distribution.Demonstrate the consistency for the following estimators.

```python
import numpy as np
import pandas as pd
mu=2.5;sigma=1.5
esp=0.05;EstP1,EstP2,EstP3=[],[],[]
n=[25,50,100,200,500,800,1000,1500,2000]
for i in n:
    x=np.random.normal(mu,sigma,(1000,i))
    EstP1.append(np.mean(abs(np.mean(x,axis=1)-mu)<esp))
    EstP2.append(np.mean(abs(np.var(x,axis=1)-sigma**2)<esp))
    EstP3.append(np.mean(abs(np.median(x,axis=1)-mu)<esp))
df=pd.DataFrame(data={'Sample Size':n,'xÃÑ':EstP1,'S^2':EstP2,'Sample Median':EstP3});df
```

```python
theta=2.5;esp=0.05;EstP1,EstP2=[],[]
n=[25,50,100,200,500,700,1000,1500,2000,2500]
for i in n:
    y=np.random.uniform(0,1,(1000,i))
    x=(1-y)**(-1/theta)
    T1=np.mean(x,axis=1)
    EstP1.append(np.mean(abs((T1/T1-1)-theta)<esp))
    T2=1/sum(np.log(x))
    EstP2.append(np.mean(abs((T2-theta)<esp)))
df=pd.DataFrame(data={'Sample Size':n,'X(1)':EstP1,'xÃÑ-1':EstP2});df
```
